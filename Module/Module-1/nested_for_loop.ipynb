Nested For Loop::
### A nested loop means one loop is inside another loop


```python
#### Example:::
for row in range(1,4):
  for col in range(1,5):
    print(f"row = {row}, and Col = {col}")
    #### Think of a grid . Outer loop = rows and inner loop = columns
```

4) Buling Output per row

```python
for row in range(1,4):
  line = ""
  for col in range(1,5):
    line += f"{col} "
    print(line)
  print("row",row,":",line)
```

5) Classic Pattern Printing (stars)

```python
#### Rectangke if stars = (3rows * 5 columns)
rows = 3
cols = 5
for r in range(1,rows):
  line = ""
  for c in range(cols):
    line += "*"
  print(line)
```

```python
### Right angle ::
rows = 5
for r in range(1,rows):
  line = ""
  for c in range(r):
    line += "*"
  print(line)
```

```python
### Inverted Triangle
rows = 5
for r in range(rows,0,-1):
  line = ""
  for c in range(r):
    line += "*"
  print(line)
```

#### Displaying the  Matrix
####  A 2d list is a list of lists. [[1,2,3],[4,5,7],[7,8,9]]


```python
##### Example :::
matrix = [[1,2,3],[4,5,6],[7,8,9],]
print(type(matrix))
print(matrix)
for row in matrix:
  for value in row:
    print(value)
    print(value,end = " ")
  print()
```

```python
### Displaying the index in list1
list1 = [1,2,34]
for i in range(len(list1)):
  print(i,list1[i])
```

```python
#### sum of the all values in a 2d list
matrix = [[1,2,3],[4,5,6],[7,8,9],]
total = 0
for row in matrix:
  for value in row:
    total += value
print(total)
```

```python
#### Multiplication of the table ::
for i in range(11):
  for j in range(10):
    print(f"{i} * {j} = {i*j}")
  print()
```

```python
#### Multiplication of the table ::
for i in range(1,11):
  for j in range(1,11):
    print(f"{i*j:4}",end = " ")
  print()
```
