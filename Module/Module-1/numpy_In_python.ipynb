Arrary Properties::

```python
### Types of the array properties :: such as .shape,,size .dtype,.ndim and the .zeros,.ones , .arange()
import numpy as np
arr = np.array([[1,2,3,9],[4,5,6,0]])
print(arr.shape) ### It show the shape of the array
print(arr.dtype) ### It show the data type of the array
print(arr.ndim) ### It show the dimension of the array
print(arr.size) ### It show the how many item  of the array are there is present
print(arr.itemsize) ### It show the size of the array
```

```python
#### The Other Remaining Properties in a array ::
import numpy as np
arr = np.zeros((3, 4)) ### it gives the 3*4 matrix all the values in the zeros
print(arr)
arr = np.ones((3, 4)) ### it gives all the values of 3*4 in the 1
print(arr)
arr1 = np.array([1,2,3])
arr = np.empty((3,3)) ### it gives the empty garbarge Value's
print(arr)
```

```python
### Example 3:: ### Scalrized method
import numpy as np
import time
arr = np.arange(1000000)
start = time.time()
for i in range(len(arr)):
  arr[i] = arr[i]*2
end = time.time()
print(end-start)
### For the Vectorized Method ::
### We don't used the for loop in the numpy ,pandas
arr = np.arange(1000000)
start = time.time()
arr = arr*2
end = time.time()
print(end-start)
```

```python
import numpy as np
arr = np.eye(3) ### it gives it identity Matrix
print(arr)
arr = np.diag((1,3)) ### It gives the diagonal which is given on a list
print(arr)
```

2. Range Based :::


```python
import numpy as np
### generating the range of the array ::
arr = np.arange(5,9) ### it work same as the range Property::
print(arr)
arr = np.arange(1,10,2)
print(arr)
#### Generating the linespace of the array::
arr = np.linspace(1,4,9) ### start = 1 , and stop = 9 and gap = 9-1 = 8 values is generating based on that line of the code
print(arr)
arr = np.linspace(1,2,3)
print(arr)
```

Stucture of the array


```python
# .reshape() it is used for the reshaping the array

# .flatten() it is used for the new copy

# .ravel() it is usef for the viewing the data structure

# .T it is used for the transporing the array
import numpy as np
arr = np.array([[1,2,5],[8,9,0]])
print(arr)
### Reshaping the Array
reshaping1 = arr.reshape(3,2)
print(reshaping1)
### New Copying the Values
copying2 = arr.flatten()
print(copying2)

### viewing the array values
viewing3 = arr.ravel()
print(viewing3)
### Transposew
result = arr.transpose() ### it transpose
result2 = arr.T
print(result2)
print(result)

```
