If Statement :: If statement: executes the code block only when the condition is true; otherwise it does nothing.

```python
age = 18
if age >= 18:
  print("You can Vote")
```

Important Indentation

If,Else Statement:: Checks multiple conditions and executes only one block whose condition is true; otherwise runs else.

```python
age = 7
if age >= 18:
  print("You can Vote")
else:
  print("You can't Vote")
```

If,elif, and else Statement:: executes the if block when the condition is true; otherwise executes the else block

```python
score = 25
if score >= 90:
  print("Grade A+")
elif score >= 80:
  print("Grade A")
elif score >= 70:
  print("Grade B+")
elif score >= 60:
  print("Grade B")
  print("You can do better")
else:
  print("Fail")
  print("Next Time")
print("Thank you ! You can do best Next Time")
```

### 1) And Operators :: Both Condition must be True otherwise False
### 2) Or Operators :: One condition must be True or satify otherwise False
### 3) Not Operators :: It Revese the Contion if True then flip to False othewise True



```python
### Example 1::
a = 10
b = 30
if a < b:
  print("a is less than b")
elif a == b:
  print("a is equal to b")
elif a>=b:
  print("a is greater than or equal to b")
else:
  print("a is greater than b")
```
