Different Method used in the set:::

```python
#### Using the Properties like add, update,remove,discard,pop,clear in the set
set1 = {1,2,5,10,50}
set2 = {"Subin Thapa","Garima Sapkota",'Babina Singh',"Rahul Adhikari",67}
#### Using the add Properties::
set1.add(40)
print(set1)
#### using the update Properties::
set1.update(set2)
print(set1)
#### Using the remove::
set1.remove("Garima Sapkota") ### through value
print(set1)
#### Using the Pop::
set1.pop()
print(set1)
### Using the discard Properity to remove the number
set1.discard("Subin Thapa")
print(set1)
#### Using the Clear Property to clear the set1 and making it's empty
set1.clear()
print(set1)


```

Different  Mathematical Operation in the Tuple::

1) Union:: For Every Element Set1 and Set2 as well as denoted by |
2) Intersection :: Common Element in Set1 and Set2 denoted by &
3) Difference :: Difference in the both the set denoted by the (-) mean only either set1 or set2
4) Symmetrical _differnece :  Either in one set or another but not in both set
it is denoted by (^)
5) subset:: a set whose elements are all contained within both of the original sets and denoted by <=


6)Superset::: any set (Set A) that contains all the elements of Set B, and potentially mean like set A is containg all in element in the set B if two set are there denoted by >=
7) Disjoint::: Set has no common Element

Number Testing on the Set Method::


```python
### Some Example of Set Method::
### 1st Work Illustration::
set1 = {"Subin Thapa","Angat Sitoula","Mahesh Basnet","Sagar Regmi"}
set2 = {"Angat Sitoula","Naresh Basenet","John Von Neumann","Pratik Manandhar","Rahul","Subin Thapa"}
### Doing Intersection,Union,Difference, and Symmetrical Differnce
result1 = set1.intersection(set2) ### Common in both set
print(result1)
result2 = set1.union(set2) ### Display the Overall Result
print(result2)
result3 = set1.difference(set2) ### Difference element in set1 from the set2
print(result3)
result4 = set1.symmetric_difference(set2) ### Two set common Element::
print(result4)
#### Other Properties Used such as intersection_update,union_update,difference_update,symmetric_differnce_update
set1.symmetric_difference_update(set2) ### it modify the set1 with the symmetric difference output::
print(set1)
#### if i want the intersection Update ouput::
set1.intersection_update(set2)
print(set1)
```

```python
### 2nd Work Illustration::
### Checking the Element:::
### superset,subset,and disjoint
set1 = {1,5,7,8}
set2 = {1,5}
result = set1.issubset(set2) ### Output Must be false::
print(result)
reselt = set2.isdisjoint(set1) ### Output must be false because every set must have unique element::
print(reselt)
result3 = set1.issuperset(set2)
print(result3)
```
