```python
### Today, I learn about the variable, Keyword, identifier, and the comment
### Soft Reserved Word::
# ## It means that it is used on both the keyword as well as the identifier as well
### There are 3 types of soft reserved word, such as match, case, and the last one is _ they also work as variable outside match statement it:
match = "Subin Thapa"
print(match)
### 2nd Example::
x = 1030
match x:
  case 1:
    print("One")
  case 100:
    print("Hundred")
  case 200:
    print("2 Hundred")
  case _:
    print("I ignore other things: ")
```
