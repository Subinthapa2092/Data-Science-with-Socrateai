### Function Calling

```python
#### Example of the Function::
num1 = [1,2,3,4,5,6]
num2 = [5,78,8]
def count(num):
    count = 0
    for i in num:
        count = count + 1
    return count
print(count(num1))
print(count(num2))
```

```python
#### Example 3:;
def check_evenodd(n):
  if n%2 == 0:
    print("Even")
  else:
    print("Odd")
check_evenodd(2)
check_evenodd(3)
```

```python
### Finding the Cube of the Number ::
def square(n):
  return n*n
def cube(n):
  return square(n)*n
print(cube(3))
```
