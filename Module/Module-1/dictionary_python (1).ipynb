3) Safe Access with get()

get() is safe::

```python
### Example 1st::
student = {
    "name" : "Subin Thapa",
    "roll_no": 34,
    "Class" : "Bachelor in Data Science",
    "Semester" : "1st Semester"
}
print(student.get("age")) ### It return None::
print(student.get("age",18)) ### It return 18 on the age but doesn't update the whole age ::
print(student)
```

9) Dictionary Concatenation / Merge in Python::
There are 2 Common Ways::

I) | (vertical bar) operator ::

II) ** Unpacking::

```python
### Working with the some Examples::
square = {1: 1, 2: 4, 3: 9, 4: 16, 5: 25}
cube = {1:1,2:8,3:27,4:64,5:125}
fourth = {1:1,2:16,3:81,4:256,5:625}
print(square | cube |fourth) ### last one Varialble win if the same keys is available::
print(square | cube) ## cube win as both have the same key
print(student |square) ### it merged the result outcome::
### another way also work such as
print(square,cube) ### It also merge the code:
```

```python
#### unpacking Some Example::
square = {1: 1, 2: 4, 3: 9, 4: 16, 5: 25}
cube = {1:1,2:8,3:27,4:64,5:125}
merged = {**square,**cube} ### Last one get priority if there are the same keys::
merged1 = {**student,**cube} ### Unpacking it also
print(merged)
print(merged1)
```

### iii) pop(key,default) -:: if key is missing,it return default (no crash)

```python
#### Some Example ::-
student = {"name":"Subin Thapa","age":18,"semester":"1st Semester","class":"Bachelor in Data Science"}
### pop(key)
result = student.pop("name")
print(result)
print(student)
### del dict["key"] dictionary key
del student["semester"]
print(student)
### pop(key,default)
### unknown key trying :
result1 = student.pop("City","default")
print(result1)
print(student)
#### popitem()
print(student.popitem())
print(student)
```

12) Sorting a Dictionary
### Dictionary keep insertion order, but sorting is done using only sorted() on items


```python
#### Example 1::
scores = {"Ram":70,"Sita" : 90,"Rakesh":78,"Gita":67}
### Sorting a dictionary with different methods::
### sort the key by (alphabetical)
sorted_by_keys = dict(sorted(scores.items()))
print(sorted_by_keys)
```

```python
### Example 2:::
scores = {"Ram":70,"Sita" : 90,"Rakesh":78,"Anita":67}
### Sort by the Values using the lambda function with expression, argument
### Displaying the hightest to lowest Marks
sorted_by_values = dict(sorted(scores.items(),key = lambda x:x[1],reverse = True))
print(sorted_by_values)
```

```python
### Example 3:::
scores = {"Ram":70,"Sita" : 10,"Rakesh":78,"Gita":67}
### Just normal sorting with revese mean decending order::
result = dict(sorted(scores.items(),reverse = True))
print(result)
```

```python
### Example 4:: Using the Last one Lamda function

scores = {"Ram":70,"Sita" : 90,"Rakesh":78,"Gita":67}

sorted_by_result = dict(sorted(scores.items(),key = lambda x:x[1]))
print(sorted_by_result)
```

14) Shallow Copy Vs Deep Copy
####
A shallow copy creates a new object, but does not create copies of the nested objects inside it. Instead, it references the original nested objects.

Changes to nested objects affect both original and copy.

Can be created using copy.copy() in Python.
#### """

Deep Copy:
A deep copy creates a completely independent copy of the object, including all nested objects.

Changes to the copy do not affect the original object.

Can be created using copy.deepcopy() in Python.
### """"

```python
import copy
even_number = {2:4,4:16,6:36,8:64}

### Shallow Copy
shallow_copy = copy.copy(even_number)
### updating ::
shallow_copy[10] = 100
print(even_number)
print(shallow_copy)
### Deep Copy
deep_copy = copy.deepcopy(even_number)
print(deep_copy)
print(even_number)
```
