```python
df["age"] = df["age"].fillna(df["age"].mean())
df.isnull().sum()
df["age"].mean()
```

```python
sns.countplot(x="survived",data = df)
plt.title("Survival Count (0= No and 1 = Yes)")
```

### More People died than Survived

```python
sns.countplot(x="sex",hue = "survived",data = df)
plt.title("Sex Count with Surival")
```

#### Female had much higher survival rate


```python
sns.countplot(x="class",hue = "survived",data = df)
plt.title("Survived by Passenger Class")
```

### 1st class Passenger survived the most
### 3rd class Passenger died the most

```python
plt.hist(df["age"],bins =20)
plt.title("Age Distribution")
plt.xlabel("Age")
plt.ylabel("Count")
plt.show()
```


### Most Passengers were young adults(20-40)
