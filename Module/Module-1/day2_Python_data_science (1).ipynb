Working with the slicing

```python
number = [1,2,3,4,5,7,9,100,200]
result = number[:4] ### start and the stop where stop is default::
print(result)
result1 = number[::2] ### start : stop: step where start and the stop is default
print(result1)
#### Dispalying the index 2 to 3 number
print(number[2:4])
```

```python
#### Clearing the whole list  in the python
### using the clear()
namelist = ["Babina",'Saloni',"Garima","Nikita"]
namelist.clear()
print(namelist)
```

```python
#### using the pop with index and without pop to delete the list items::

namelist1 = ["Rahul","Mahesh","kiran",[23,24,256],{56,89,90}]
namelist1.pop() ### this remove last items::
namelist1.pop(0) ### this remove with the indexing number::
print(namelist1)
```

```python
#### using the remove method to delete the items using the list items::
namelist = ["Subin Thapa","Angat Sitoula",18,19,"Rahul Adhikari","Subin Thapa"]
namelist.remove("Subin Thapa") #### it removes only the first occurence of the item if there are the multiple occurence or duplicate
print(namelist)
```

```python
### Ascending Order displaying using sort
student_list = ["Subin Thapa","Pratik Sharma","Madhav Khatri","Sagar regime","Aman"]
student_list.sort()
print(student_list)
### Displaying the acending number
number = [12,13,4,3,2,90,100,1,79]
result = sorted(number)
print(result)
```
